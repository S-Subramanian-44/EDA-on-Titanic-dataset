```python
df = df.fillna(df['Age'].mean())
```

```python
df.isna().sum()
```

```python
(df.Survived.value_counts() / df.shape[0] * 100).plot.bar(title='Target distribution')
```

0 --> Not Survived <br>
1 --> Survived <br>
From the plot we can see that the survival rate is less

```python
plt.figure(figsize=(10, 5))
plt.title('Frequency distribution of Age')
sns.histplot(df['Age'])
plt.show()
```

Distibution of people around the age of 20-40 are more in the titanic ship

```python
plt.figure(figsize=(10, 5))
plt.title('Frequency distribution of Sex')
sns.histplot(df['Sex'])
plt.show()
```

The are more male than female in the titanic ship

```python
plt.figure(figsize=[8,5])
sns.countplot(x="Sex",hue="Survived",data=df)
```

The survival rate of male is less when compared to female

```python
plt.figure(figsize=[8,5])
sns.countplot(x="SibSp",hue="Survived",data=df)
```

The passengers with no sibilings have high survival rate

```python
plt.figure(figsize=[8,5])
sns.countplot(x="Pclass",hue="Survived",data=df)
```

Paasengers on class 1 has more survival rate than the passengers on class 2 and 3

```python
plt.figure(figsize=[8,5])
sns.countplot(x="Parch",hue="Survived",data=df)
```

```python
surv = sns.FacetGrid(df,col = "Survived")
surv.map(sns.histplot,"Age",bins = 25)
plt.show()
```
